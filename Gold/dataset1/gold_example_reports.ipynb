```python
from pyspark.sql.functions import col, count as spark_count, sum as spark_sum
```

```python
df_gold_models = (
    spark.table("silver.complaints")
        .groupBy("model_year", "make", "model")
        .count()
        .withColumnRenamed("count", "complaint_count")
)
```

```python
df_gold_models.write.format("delta").mode("overwrite").saveAsTable("gold.model_complaint_stats")
```

```python
df_gold_components = (
    spark.table("silver.complaint_components")
        .groupBy("component")
        .count()
        .withColumnRenamed("count", "complaint_count")
        .orderBy(col("complaint_count").desc())
)
```

```python
df_gold_components.write.format("delta").mode("overwrite").saveAsTable("gold.component_stats")
```

```python
df_gold_safety = (
    spark.table("silver.complaints")
        .groupBy("model_year", "make", "model")
        .agg(
            sum(col("crash").cast("int")).alias("total_crashes"),
            sum(col("fire").cast("int")).alias("total_fires"),
        )
)
```

```python
df_gold_safety.write.format("delta").mode("overwrite").saveAsTable("gold.safety")
```
