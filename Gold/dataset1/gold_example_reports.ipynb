```python
df_gold_models = (
    spark.table(f"{silver_schema}.complaints")
        .groupBy("model_year", "make", "model")
        .count()
        .withColumnRenamed("count", "complaint_count")
)
```

```python
df_gold_models.write.format("delta").mode("overwrite").saveAsTable(f"{gold_schema}.model_complaint_stats")
```

```python
df_gold_components = (
    spark.table(f"{silver_schema}.complaint_components")
        .groupBy("component")
        .count()
        .withColumnRenamed("count", "complaint_count")
        .orderBy(col("complaint_count").desc())
)
```

```python
df_gold_components.write.format("delta").mode("overwrite").saveAsTable(f"{gold_schema}.component_stats")
```

```python

df_gold_safety = (
    spark.table(f"{silver_schema}.complaints")
        .groupBy("model_year", "make", "model")
        .agg(
            sum(col("crash").cast("int")).alias("total_crashes"),
            sum(col("fire").cast("int")).alias("total_fires"),
        )
)
```

```python
df_gold_safety.write.format("delta").mode("overwrite").saveAsTable(f"{gold_schema}.safety")
```
